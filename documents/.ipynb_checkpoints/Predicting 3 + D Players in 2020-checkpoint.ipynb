### 1.1 Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#three point predictions
df_three = pd.read_csv('C:\\Users\\David\\OneDrive\\Projects\\nba-draft-player-analysis\\predictions\\results_2020_three_point.csv')

#three point predictions
df_def = pd.read_csv('C:\\Users\\David\\OneDrive\\Projects\\nba-draft-player-analysis\\predictions\\results_2020_defense.csv')

#merge results
df_final = pd.merge(df_three[['name','three_point_attempts_per_40','predict']].rename(columns={'predict':'three_point_predict'})
                    ,df_def[['name','steals_per_40_predict','blocks_per_40_predict']],how='inner',on='name')

df_final.head()
```

### 1.2 Steals vs. Three Point %

```python
sns.set_style("whitegrid", {
    "ytick.major.size": 0.1,
    "ytick.minor.size": 0.05,
    'grid.linestyle': '--'
 })

ax = sns.lmplot('three_point_predict', # Horizontal axis
           'steals_per_40_predict', # Vertical axis
           data=df_final, # Data source
           fit_reg=False, # Don't fix a regression line
           size = 8,
           aspect =2) # size and dimension

plt.title('Steals per 40 vs. Three Point %')
# Set x-axis label
plt.xlabel('Three Point %')
# Set y-axis label
plt.ylabel('Steals per 40')


def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x'], point['y'], point['val'])

label_point(df_final['three_point_predict'], df_final['steals_per_40_predict'], df_final['name'], plt.gca()) 
```

### 1.3 Stocks vs. Three Point %

```python
df_final['stocks_per_40_predict'] = df_final['steals_per_40_predict']+df_final['blocks_per_40_predict']

sns.set_style("whitegrid", {
    "ytick.major.size": 0.1,
    "ytick.minor.size": 0.05,
    'grid.linestyle': '--'
 })

ax = sns.lmplot('three_point_predict', # Horizontal axis
           'stocks_per_40_predict', # Vertical axis
           data=df_final, # Data source
           fit_reg=False, # Don't fix a regression line
           size = 8,
           aspect =2) # size and dimension

plt.title('Stocks per 40 vs. Three Point %')
# Set x-axis label
plt.xlabel('Three Point %')
# Set y-axis label
plt.ylabel('Stocks per 40')


def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x'], point['y'], point['val'])

label_point(df_final['three_point_predict'], df_final['stocks_per_40_predict'], df_final['name'], plt.gca()) 
```

### 1.4 Three Point % vs. Three Point Attempts

```python
names = ['Trevelin Queen','Tyrese Haliburton','Devin Vassell','Justinian Jessup','Tyrese Haliburton','Tyrell Terry','Payton Pritchard',
         'Desmond Bane']
df_top = df_final[df_final['name'].isin(names)]

sns.set_style("whitegrid", {
    "ytick.major.size": 0.1,
    "ytick.minor.size": 0.05,
    'grid.linestyle': '--'
 })

ax = sns.lmplot('three_point_predict', # Horizontal axis
           'three_point_attempts_per_40', # Vertical axis
           data=df_top, # Data source
           fit_reg=False, # Don't fix a regression line
           size = 8,
           aspect =2) # size and dimension

plt.title('Three Point % vs. Three Point Attempts')
# Set x-axis label
plt.xlabel('Three Point %')
# Set y-axis label
plt.ylabel('Three Point Attempts')


def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x'], point['y'], point['val'])

label_point(df_top['three_point_predict'], df_top['three_point_attempts_per_40'], df_top['name'], plt.gca()) 
```
