## 1.0 | Exploratory Data Analysis

The objective of this EDA is to take a deeper dive into what is predictive of NBA 3pt % in the first 4 years of a player's career. Like with predicting NBA VORP, looking at values overa the first 4 years does a few things. First, it allows for a large sample of data. If it was career 3pt %, then all current players would need to be excluded since their careers are unfinished. Second, 4 years is the length of a rookie contract, so this is the best timespan to approximate the value a team is getting back on a drafted player before they become a free agent.

#### 1.2 NBA 3pt Shooting Correlations

Suprisingly, college 3pt % shows a weak correlation with nba 3pt %. In fact, college three from % seems to show a better correlation with nba 3pt %. The actual number of 3pt attempts is the strongest correlated variable. I would expect there is some unaccounted variance around how many shots a player takes in college to acheive their 3pt %. I would have a lot more faith in a 40% 3pt shooter over 100 attempts in college than a 50% shooter over 30 attempts. Additionally, not every player is going to have their best results the final season of college. Some players will likely get injured and post incomplete statistics. 

Combining college 3pt attempts and college 3pt % gives us a better approximation of nba 3pt % by weighting the value with the number of attempts

```python
corr_fields = ['free_throw_percentage','free_throw_attempts_per_40','three_point_percentage',
               'three_point_attempts_per_40','nba_three_point_percentage']

for i in [corr_fields]:
    corr_chart(df_corr = df[i])
```
