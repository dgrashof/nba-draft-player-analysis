### 1.1 | College Basketball

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cc_df = pd.read_csv('data/cc_df.csv')
#check for nulls
cc_df.groupby('season')['season'].agg(len)
cc_df = cc_df[~cc_df.season.isin(['2004-05','2005-06','2006-07','2007-08'])]
cc_df_career = cc_df[cc_df['season']=='Career']
cc_df = cc_df[cc_df['season']!='Career']
cc_df = cc_df.drop(columns = 'Unnamed: 0')
```

```python
#label years played
cc_df = cc_df.sort_values(by=['player_id','season'],ascending = True)
cc_df['yrs_played'] = cc_df.groupby('player_id').cumcount()+1
exclude_list = [i for i in cc_df['player_id'][cc_df['season']=='2008-09']]
cc_df = cc_df[~cc_df['player_id'].isin(exclude_list)]
cc_df = cc_df.dropna()
```

### 1.1.1 | Player Height

__Hypothesis__: The taller a player is for their position, the higher their box score plus minus
<br>
<br>
__Result__: There does not appear to be much correlation between height and college production by position or otherwise. A follow-up question would be whether this trend applies to NBA production as well. 

```python
#convert height into integer and all inches
cc_df['feet'] = cc_df['height'].str[:1]
cc_df['inches'] = cc_df['height'].str[2:]
cc_df['feet'] = cc_df['feet'].astype(int)
cc_df['inches'] = cc_df['inches'].astype(int)
cc_df['height_revised'] = (cc_df['feet']*12)+cc_df['inches']
cc_df = cc_df.drop(columns = 'height')
cc_df = cc_df.rename(columns = {'height_revised':'height'})
```

```python
#all players
plt.scatter(cc_df['height'],cc_df['box_plus_minus'])
z = np.polyfit(cc_df['height'], cc_df['box_plus_minus'], 1)
p = np.poly1d(z)
plt.plot(cc_df['height'],p(cc_df['height']),"r--")
plt.xticks(rotation = 90)
plt.title('Height versus BPM')
plt.show()
```

```python
#players by position
for i in set(cc_df['position']):
    plt.scatter(cc_df['height'][cc_df['position']==i],cc_df['box_plus_minus'][cc_df['position']==i])
    z = np.polyfit(cc_df['height'][cc_df['position']==i], cc_df['box_plus_minus'][cc_df['position']==i], 1)
    p = np.poly1d(z)
    plt.plot(cc_df['height'][cc_df['position']==i],p(cc_df['height'][cc_df['position']==i]),"r--")
    plt.xticks(rotation = 90)
    plt.title('Height versus BPM by Position: '+i)
    plt.show()
```
