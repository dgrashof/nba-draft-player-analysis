```python
def optimize_beta(value,weight,upper,lower,alpha_range,beta_range):
    idx=0
    results = pd.DataFrame()
    for i in alpha_range:
        for j in beta_range:
            new_value = ((weight*value)+i)/(weight+i+j)
            results = results.append(pd.DataFrame({'Alpha':i,'Beta':j,'Orig_mean':np.mean(value),'Orig_min':lower,'Orig_max':upper,
                                      "Revised_mean":(1+i)/(1+i+j),"Revised_min":np.min(new_value),
                                      "Revised_max":np.max(new_value)},index=[idx]))
            idx = idx+1
    results['min_diff'] = np.abs(results['Orig_min']-results['Revised_min'])
    results['max_diff'] = np.abs(results['Orig_max']-results['Revised_max'])
    results['mean_diff'] = np.abs(results['Orig_mean']-results['Revised_mean'])
    results['min_total'] = np.min(results['mean_diff']+results['max_diff']+results['min_diff'])
    results = results.sort_values(by='mean_diff',ascending=True)[:50]
    results = results.sort_values(by='min_diff',ascending=True)[:25]
    results = results[results['max_diff']==np.min(results['max_diff'])]
    return(results['Alpha'].values,results['Beta'].values)
```

```python
alpha,beta = optimize_beta(np.array(df['nba_three_point_percentage']),np.array(df['nba_attempted_three_point_field_goals'])
              ,.50,.20,range(1,300),range(1,300))

x = np.linspace(bt.ppf(0.01, alpha, beta),
                bt.ppf(0.99, alpha, beta), 100)

rv = bt(alpha, beta)
plt.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
plt.title('Selected Beta Distribution - Alpha:'+str(alpha)+' Beta:'+str(beta))
```

```python
#create revised nba 3pt percentage
df['revised_nba_three_point_percentage'] = (df['nba_three_point_made']+alpha)/(df['nba_attempted_three_point_field_goals']+alpha+beta)
df[['col_lkup','nba_three_point_made','nba_attempted_three_point_field_goals','nba_three_point_percentage',
    'revised_nba_three_point_percentage']].sort_values(ascending=False,by='revised_nba_three_point_percentage')
```

### 1.4 Create Regression - Position Specific Alpha and Betas

```python
for i in ['Center','Forward','Guard']:
    df['nba_three_point_percentage'][df['position']==i].hist()
    plt.title('Position:'+i)
    plt.show()
```

```python
def apply_ab(column_num,column_den,ab):
    new_column = (column_num+ab[0])/(column_den+ab[0]+ab[1])
    return(new_column)

df_rev = pd.DataFrame()

for pos,ab,cols in zip(['Forward','Guard','Center'],[(24,51),(24,51),(4,31)],[('nba_three_point_made','nba_attempted_three_point_field_goals')]*3):
    temp = df[df['position']==pos]
    temp['rev_nba_three_point_percentage']=apply_ab(np.array(temp[cols[0]]),np.array(temp[cols[1]]),ab)
    df_rev = df_rev.append(temp)

df_rev_ii = pd.DataFrame()

for pos,ab,cols in zip(['Forward','Guard','Center'],[(21,39),(21,39),(16,52)],[('three_point_made','three_point_attempts')]*3):
    temp = df[df['position']==pos]
    temp['rev_three_point_percentage']=apply_ab(np.array(temp[cols[0]]),np.array(temp[cols[1]]),ab)
    df_rev_ii = df_rev_ii.append(temp)
    
df_rev['rev_three_point_percentage'] = df_rev_ii['rev_three_point_percentage']

df_rev_iii = pd.DataFrame()

for pos,ab,cols in zip(['Forward','Guard','Center'],[(211,68),(211,68),(214,110)],[('free_throws_made','free_throw_attempts')]*3):
    temp = df[df['position']==pos]
    temp['rev_free_throw_percentage']=apply_ab(np.array(temp[cols[0]]),np.array(temp[cols[1]]),ab)
    df_rev_iii = df_rev_iii.append(temp)
    
df_rev['rev_free_throw_percentage'] = df_rev_iii['rev_free_throw_percentage']
```
